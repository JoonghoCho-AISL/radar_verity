```python
import tensorflow as tf
from MobiusAPI import http_post_get
import json
import numpy as np
gpus = tf.config.list_physical_devices('GPU')
if gpus:
  # 텐서플로가 첫 번째 GPU만 사용하도록 제한
  try:
    tf.config.set_visible_devices(gpus[2], 'GPU')
  except RuntimeError as e:
    # 프로그램 시작시에 접근 가능한 장치가 설정되어야만 합니다
    print(e)
```

```python
model = tf.keras.models.load_model('model/resnet_pool_disk_split')
```

```python
URI = '/Mobius/Joongho/radar'
AE_ID = 'Joongho'
```

```python
data = http_post_get.mobius_get(URI, AE_ID, AE_ID)
data = json.loads(data)
radar_data = data['m2m:cin']['con']
radar_data = np.array(radar_data)
y = model.predict(radar_data)
print(y)
print(np.argmax(y))
```
